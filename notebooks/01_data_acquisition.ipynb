```python
# 01_data_acquisition.ipynb
# Purpose: Load the UCI Concrete Compressive Strength dataset,
# inspect the raw data, clean column names, and save a processed CSV.

import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Define file paths
data_dir = Path("../data")
raw_file = data_dir / "concrete+compressive+strength" / "Concrete_Data.xls"

# Load the raw Excel dataset
df = pd.read_excel(raw_file)

df.head()
```

```python
print("Dataset shape:", df.shape)

print("\nMissing values:")
print(df.isnull().sum())
```

```python
df.columns = [
    "cement",
    "slag",
    "fly_ash",
    "water",
    "superplasticizer",
    "coarse_aggregate",
    "fine_aggregate",
    "age",
    "compressive_strength"
]

df.head()
```

```python
df["w_c_ratio"] = df["water"] / df["cement"]

df[["water", "cement", "w_c_ratio"]].head()
```

```python
import numpy as np

df["log_age"] = np.log(df["age"])

df[["age", "log_age"]].head()
```

```python
df.info()
```

```python
processed_file = "../data/concrete_processed.csv"

df.to_csv(processed_file, index=False)

print("Saved:", processed_file)
```

## Summary

*This notebook loaded the UCI Concrete Compressive Strength dataset, inspected the raw data, confirmed that the dataset contains 1,030 rows and no missing values, renamed the columns for easier analysis, and created two engineered features based on instructor feedback: water-cement ratio (`w_c_ratio`) and logarithmic curing age (`log_age`). The processed dataset was saved as `data/concrete_processed.csv` for use in the EDA and modeling notebooks.*
